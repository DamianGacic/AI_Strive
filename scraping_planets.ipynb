```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Replace this with the HTML code for the table body element
url = "https://in-the-sky.org/ephemeris.php?ird=1&irs=1&ima=1&iob=1&objtype=1&objpl=Mars&objtxt=Mars&tz=0&startday=22&startmonth=9&startyear=2023&interval=4&rows=25"  # Replace with your URL

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

# Find all rows (tr) in the table
rows = soup.find_all('tr')

# Initialize lists to store data
data = []
columns = ['Year', 'Month', 'Day', 'Time', 'Moon Age', 'Right Ascension', 'Declination', 'Time 1', 'Time 2', 'Time 3', 'Magnitude', 'Observable', 'Constellation']

# Extract data from each row
for row in rows:
    # Extract text from each cell (td) in the row
    cells = row.find_all('td')
    row_data = [cell.get_text(strip=True) for cell in cells]
    
    # Append the row data to the list
    data.append(row_data)

# Create a pandas DataFrame from the data
df = pd.DataFrame(data, columns=columns)


```
